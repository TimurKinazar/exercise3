```python
#1
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('populations.txt')
year, hare, lynx, carrot = data.T
plt.axes([0, 0, 1, 1]) 
plt.plot(year, hare,year, lynx, year, carrot) 
plt.legend(('Hare', 'Lynx', 'Carrot'), loc=(1.04, 0.4))
plt.xlabel(u'Годы')
plt.ylabel(u'Численность')
plt.show()
```

```python
#2
import pandas as pd
df=pd.DataFrame({'hare':hare,'lynx':lynx})
totals=[i+j for i,j in zip(df['hare'],df['lynx'])]
hareBars=[i/j *100 for i,j in zip(df['hare'],totals)]
lynxBars=[i/j *100 for i,j in zip(df['lynx'],totals)]
plt.bar(year,hareBars, color='#b5ffb9',edgecolor='white',width=1)
plt.bar(year,lynxBars,bottom=hareBars, color='#f9bc86',)
plt.legend(('Hare', 'Lynx'), loc=(1.04, 0.4))
plt.show()
```

```python
#3
Lmean=int(lynx.mean(axis=0))
Hmean=int(hare.mean(axis=0))
Cmean=int(carrot.mean(axis=0))
print("Lynx", Lmean,"and its std - ",int(lynx.std(axis=0)))
plt.bar(year,lynx,color='#ffcc00')
plt.plot(year,np.repeat(Lmean,21), 'r-', linewidth=2.5) 
plt.legend(( 'Mean','Lynx'), loc=(1.04, 0.4))
plt.show()
print("Hare", Hmean,"and its std - ",int(hare.std(axis=0)))
plt.bar(year,hare,color='#99ccff')
plt.plot(year,np.repeat(Hmean,21), 'r-', linewidth=2.5) 
plt.legend(( 'Mean','Hare'), loc=(1.04, 0.4))
plt.show()
print("Carrot", Cmean,"and its std - ",int(carrot.std(axis=0)))
plt.bar(year,carrot,color='#ff944d')
plt.plot(year,np.repeat(Cmean,21),'g-', linewidth=2.5 ) 
plt.legend(( 'Mean','Carrot'), loc=(1.04, 0.4))
plt.show()
```

```python
#4
Hyear=np.argmax(hare)
Lyear=np.argmax(lynx)
Cyear=np.argmax(carrot)
print("В ",int(year[Hyear]),"Maximum of Hares")
print("В ",int(year[Lyear]),"Maximum of Lynxes")
print("В ",int(year[Cyear]),"Maximum of Carrots")
plt.plot(year, hare,year, lynx, year, carrot) 
plt.scatter([year[Hyear],year[Lyear],year[Cyear]],[hare[Hyear],lynx[Lyear],carrot[Cyear]], color='r')
plt.legend(('Hare', 'Lynx', 'Carrot','Maximums'), loc=(1.04, 0.4))
plt.show()
```

```python
#5
r=np.repeat('H',21)
m=np.argmax(np.array((hare,lynx)),0)
r[m==1]='L'
print(r)
labels = ['Hare','Lynx']
k=np.sum(m)
x=[21-k,k]
explode = [0, 0.05]
plt.pie(x, labels = labels,explode = explode,  autopct = '%1.1f%%', shadow=True);
plt.show()
```

```python
#6
H=hare[hare>50e3]
Hyear=year[hare>50e3]
L=lynx[lynx>50e3]
Lyear=year[lynx>50e3]
C=carrot[carrot>50e3]
Cyear=year[carrot>50e3]
print("Hares > 50000: ", Hyear)
print("Lynxes > 50000: ", Lyear)
print("Carrots > 50000 : ", Cyear)
plt.plot(year, hare,year, lynx, year, carrot) 
plt.scatter(Hyear,H)
plt.scatter(Lyear, L)
plt.scatter(Cyear,C)
plt.legend(('Hare', 'Lynx', 'Carrot','Hares > 50000','Lynxes > 50000','Carrots > 50000'), loc=(1.04, 0.4))
plt.show()
```

```python
#7
Hyear=np.argsort(hare)
Lyear=np.argsort(lynx)
Cyear=np.argsort(carrot)
print("В ",int(year[Hyear[0]])," и ", int(year[Hyear[1]]), "are lowest for Hare")
print("В ",int(year[Lyear[0]])," и ", int(year[Lyear[1]]), "are lowest for Lynx")
print("В ",int(year[Cyear[0]])," и ", int(year[Cyear[1]]), "are lowest for Carrots")
```

```python
#8
H=np.gradient(hare,1.0)
plt.plot(year,H,year, lynx)
plt.legend(('Hare change', 'Lynx change'), loc=(1.04, 0.4))
plt.xlabel(u'Year')
plt.ylabel(u'Population')
plt.show()
print("коэффициент корреляции ",np.corrcoef(H,lynx)[0,1])
```

```python
## 10
import pandas as pd
import cmath 
hare_grad = np.gradient(hare, 1.0)
maxHare=np.max(hare)
minHare=np.min(hare)
maxHare_grad=np.max(hare_grad)
minHare_grad=np.min(hare_grad)
maxLynx=np.max(lynx)
maxCarrot=np.max(carrot)
print("                      Hares  ")
plt.plot(year,hare_grad)
plt.show()
print("             closest sin graph to 'Hares'")
arc=cmath.asin((hare_grad[0]-minHare_grad)/(maxHare_grad-minHare_grad))*np.pi
x=np.linspace(-np.pi+arc, 3*np.pi+arc, 21, endpoint=True)
S=np.sin(x)*((maxHare_grad-minHare_grad)/2)+(maxHare_grad+minHare_grad)/2
plt.plot(x,S)
plt.show()

hare_fut=int(hare[19]+S[20])
print("assumed number of hares in 1920: ",hare_fut)
print("error : ",np.abs(hare_fut-int(hare[20])))
print ("ratio of the assumed value to the true value is : " , hare_fut/24.7e3)
print ("ratio of the error to the true value is : ", np.abs(hare_fut-int(hare[20]))/24.7e3)
```
